a. creare un dataframe contenente le informazioni riportate qui sotto, relative a quanti prodotti ha
venduto (‘quantity’) e quanto ha venduto in termini di $ (‘total_amount’) ogni membro dello staff
(colonna “last_name”):

```python
ammontare = {'last_name':['Boyer','Copeland','Daniel','Serrano','Terrell','Vargas'],
          'quantity':[766,213,812,269,100,114],
          'total_amount':[643049.37,203549.39,744259.20,253183.68,93851.95,123156.90]}
```

```python
import pandas as pd
quantita = pd.DataFrame(ammontare)
quantita
```

b. visualizzi due grafici a barre affiancati che indichino quantità vendute e valore venduto per ogni
dipendente.

```python
x1 = list(quantita['last_name'])
y1 = list(quantita['quantity'])
y2 = list(quantita['total_amount'])

fig=plt.figure(figsize=(18,8))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

plt.rcParams.update({'font.size':9})
ax1.bar(x1,y1)
ax2.bar(x1,y2)
```

c. visualizzi i grafici precedenti con un titolo, gli assi e la griglia.

```python
x1 = list(quantita['last_name'])
y1 = list(quantita['quantity'])
y2 = list(quantita['total_amount'])

fig = plt.figure(figsize=(18,10))
ax1 = plt.subplot(1,2,1)
ax2 = plt.subplot(1,2,2)

plt.rcParams.update({'font.size':9})
ax.bar(x1,y1)
ax.bar(x1,y2)
ax1.set_xlabel('commesso')
ax1.set_ylabel('quantita venduta')
ax2.set_ylabel('valore')
ax2.set_xlabel('commesso')
fig.suptitle('numeri di ordini')
```

```python
fig = plt.figure(figsize=(7,8))
ax = plt.axes()

ax.bar(quantita['last_name'],quantita['total_amount']/quantita['quantity'],color='c')
ax.set_xlabel('commessi')
ax.set_ylabel('costo medio')
fig.suptitle('costo medio che viene venduta da ogni commesso')
```
