```python
provincie = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Geopandas Esercizi/Data/Limiti01012021_g/ProvCM01012021_g')
provincie
```

```python
regioni = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Geopandas Esercizi/Data/Limiti01012021_g/Reg01012021_g')
regioni
```

```python
comuni = geopandas.read_file('/content/drive/MyDrive/Colab Notebooks/Geopandas Esercizi/Data/Limiti01012021_g/Com01012021_g')
comuni
```

visualizzare il nome della provincia e il nome della regione

```python
Prov_reg = pd.merge(provincie,regioni,how='inner',on='COD_REG')[['DEN_UTS','DEN_REG']]
Prov_reg
```

visualizzare il comune e in quale regione si trova

```python
Com_Reg = pd.merge(comuni,regioni,how='inner',on='COD_REG')[['COMUNE','DEN_REG']]
Com_Reg
```

numero di comuni per ogni regione

```python
Num_Com_Per_Reg = pd.merge(comuni,regioni,how='inner',on='COD_REG').groupby('DEN_REG')['COMUNE'].count().sort_values()
Num_Com_Per_Reg
```

numero di comuni per ogni provincia

```python
Num_Com_Per_Prov = pd.merge(comuni,provincie,how='inner',on='COD_PROV').groupby('DEN_UTS')['COMUNE'].count().sort_values()
Num_Com_Per_Prov
```

numero di provincie per ogni regione

```python
Num_Prov_Per_Reg = pd.merge(provincie,regioni,how='inner',on='COD_REG').groupby('DEN_REG')['DEN_UTS'].count().sort_values()
Num_Prov_Per_Reg
```
