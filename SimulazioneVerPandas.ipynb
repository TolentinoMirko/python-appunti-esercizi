Fila A 9.visualizzare i film (tutte le colonne) il cui titolo comincia con la lettera W


```python
movies = movies[movies['Film'].str.startswith('W'),'Genre']
movies
```

Altro questo serve per i film che finiscono con quella lettera

```python
movies = movies[movies['Film'].str.endswith('s')]
movies
```
