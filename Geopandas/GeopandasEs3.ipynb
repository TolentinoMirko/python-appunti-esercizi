```
!pip install contextily
import contextily as ctx

```

```
ax = Quartieri.to_crs(3857).plot(edgecolor='k',facecolor='none',figsize=(15,9))
ax = Ciclabili.to_crs(3857).plot(ax=ax,edgecolor='blue',facecolor='none',figsize=(15,9))
ctx.add_basemap(ax)
```

```
# calcolare la dimensione media dell'area dei quartieri di Milano
Quartieri.Shape_Area.mean()
```

```
# visualizzare i nomi dei quartieri più grandi della media
Quartieri[Quartieri['Shape_Area']>=Quartieri.Shape_Area.mean()]['NIL']
```

```
# visualizzare i quartieri più grandi della media sulla mappa di Milano
```

```
PiuGrandi = Quartieri[Quartieri['Shape_Area']>=Quartieri.Shape_Area.mean()]
ax = PiuGrandi.to_crs(3857).plot(edgecolor='k',facecolor='none',figsize=(15,9))
ctx.add_basemap(ax)
```

```
# visualizzare una mappa con la legenda in modo che i quartieri più grandi abbiano colori diversi rispetto ai quartieri più piccoli. 
# La scala dei colori deve essere rossa e si deve riuscire ad intravedere in trasparenza la mappa sottostante
```

```
ax = PiuGrandi.to_crs(3857).plot(edgecolor='k',column='Shape_Area',legend=True,figsize=(15,9),cmap='Reds')
ctx.add_basemap(ax)
```

```
# salvare in un geodataframe le informazioni sul quartiere di Greco, prendendole dal geodataframe dei quartieri
```

```
Greco = Quartieri[Quartieri['NIL']=='GRECO - SEGNANO']
Greco
```

```
# trovare le piste ciclabili che passano per il quartiere di Greco
```

```
Ciclabili[Ciclabili.intersects(Greco.geometry.squeeze())]
```

```
# visualizzare su una mappa il quartiere di Greco e le piste ciclabili che lo attraversano
# evidenziare bene in nero le piste ciclabili 
```

```
ax = Ciclabili[Ciclabili.intersects(Greco.geometry.squeeze())].to_crs(3857).buffer(10).plot(marker=6,figsize=(10,17),facecolor='k')
Quartieri[Quartieri['NIL']=='GRECO - SEGNANO'].to_crs(3857).plot(ax=ax,facecolor='none',edgecolor='k')
ctx.add_basemap(ax)
```
